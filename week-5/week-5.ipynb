```python
from tkinter import messagebox
from tkinter import *

def button_click():
  messagebox.showinfo("Info", "Welcome to COS 102 GUI App \n")
  result = messagebox.askyesno("confirmation", "Do you want to continue")

  if result == 1:
    print("she wishes to continue")
  else:
    root.destroy()

root = Tk()
root.title("Home page")
root.geometry("300x100")

label = Label(root, text = 'Hello friend \n')
label.pack()

button = Button(root, text = 'Click Me \n', command= button_click)
button.pack()

button.config(fg = 'red', bg = 'yellow')

root.mainloop()
```
